7. Now to make the data filtered with specically Male Gender, we can implement the following line of code. This creates a new variable which contains data for only the male gender

```python
male_heartattack_data = df[df["Gender"] == "Male"]
```

8. Display the updated data to make sure it is working as needed. To see the output, make sure to run the code

```python
print(male_heartattack_data)
```

9. Now we can take our updated male data, and filter it once more by using the following code. This will take us to our final product of our filtered data which consists of **Age**, **Cholesterol Levels**, **Smoking Status**, and **Diabetes Status**. Implement the following code into a new code cell.


```python
heartattack_subset = male_heartattack_data[['Gender', 'Age', 'Cholesterol (mg/dL)', 'Smoking Status', 'Has Diabetes']]
```

10. Display the finalized subset to make sure it is working as intended. Be sure to run the code to see the final output.

```python
print(heartattack_subset)
```
